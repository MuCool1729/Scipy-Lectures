````python
## Prime number sieve

numbers = np.ones(1000, dtype=bool)
numbers[:2] = False

for i in range(2, int(np.sqrt(len(numbers))) +1):
    if not numbers[i]:
        continue
    numbers[i*i::i] = False

np.nonzero(numbers)
````

Exercise: Reductions

- Given there is a sum, what other function might you expect to see?

- What is the difference between `sum` and `cumsum`?

````python
#sum gives a single number, the sum of all elements of the collection. On the other hand,
#cumsum produces a collection of same length, with each element being the cumulative sum of all the 
#elements before it in the collection.

np.arange(10).sum()
np.arange(10).cumsum()
````
